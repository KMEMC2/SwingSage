```python
from googleapiclient.discovery import build
import os
import yt_dlp

# Set up YouTube API
api_key = 'AIzaSyC05nHx4TjWzEOZLcImUiN_wiWOO_i3rak'
youtube = build('youtube', 'v3', developerKey=api_key)

# Function to search for YouTube videos
def search_videos(query, max_results=100):
    request = youtube.search().list(
        q=query,
        part='snippet',
        type='video',
        maxResults=max_results
    )
    response = request.execute()
    return response['items']

# Function to download videos using yt-dlp
def download_videos(video_urls, download_path):
    ydl_opts = {
        'outtmpl': os.path.join(download_path, '%(title)s.%(ext)s'),  # Save with video title
        'format': 'best'  # Download best quality
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        for url in video_urls:
            try:
                print(f"Downloading {url}")
                ydl.download([url])
            except Exception as e:
                print(f"Error downloading {url}: {e}")

# Search query (e.g., "Federer backhand")
search_query = "Federer backhand"
search_results = search_videos(search_query)

# Collect video URLs based on search results
video_urls = [f"https://www.youtube.com/watch?v={item['id']['videoId']}" for item in search_results]

# Specify the path in Google Drive where you want to save the videos
download_path = '/content/drive/MyDrive/YoutubeDownloads'

# Create the directory if it doesn't exist
os.makedirs(download_path, exist_ok=True)

# Download the videos from the search results
download_videos(video_urls, download_path)

# Print the video URLs that were downloaded
print("Downloaded video URLs:")
for url in video_urls:
    print(url)
```
